## Question 2-plot

```python
import matplotlib.pyplot as plt

bptt_lengths = [20,30,40,50,60,70]
test_lstm =[148.6204135975389, 118.17177416820937, 111.5067037852555, 110.44923501312194, 108.45050273349982, 108.02259143780759]
test_transformer =[211.5667003978272, 173.2516606655956, 169.60505523467748, 166.7586521581105, 159.93409374123877, 156.69049613899017]

plt.figure(figsize=(10, 6))
plt.plot(bptt_lengths, test_lstm, label='LSTM', marker='o')
plt.plot(bptt_lengths, test_transformer, label='Transformer', marker='x')
plt.xlabel('BPTT Length')
plt.ylabel('Test PPL')
plt.title('Test PPLs for LSTM and Transformer Models at Various BPTT Lengths')
plt.legend()
plt.grid(True)
plt.savefig('q2.png')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

test_ppls = {(200, 1): 140.50513441301825, (200, 2): 132.96445987074708, (200, 3): 112.15775030858669, (400, 1): 111.35227580455822, (400, 2): 109.96663227764775, (400, 3): 109.25170922079712, (600, 1): 109.14359594976068, (600, 2): 109.14033829027623, (600, 3): 109.14033829027623}
nhid_values = [200, 400, 600]
nlayers_values = [1, 2, 3]
matrix = np.zeros((len(nhid_values),len(nlayers_values)))

for i in range(len(nhid_values)):
  for j in range(len(nlayers_values)):
    matrix[i,j] = test_ppls[(nhid_values[i],nlayers_values[j])]
c = plt.imshow(matrix, cmap='viridis')
plt.colorbar(c)
plt.xlabel('Num of Layers')
plt.ylabel('Hidden Dimension Size')
plt.xticks(np.arange(len(nlayers_values)),nlayers_values)
plt.yticks(np.arange(len(nhid_values)),nhid_values)

plt.title('Test PPL for different hidden dimension and num of layers')
plt.savefig('question3.png')
plt.show()
```
